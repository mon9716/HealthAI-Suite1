```python
# --- Imports ---
import os, random, collections, json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from google.colab import drive

# --- Mount Drive ---
drive.mount("/content/drive", force_remount=True)

# --- Paths ---
DATA_DIR = "/content/drive/MyDrive/HealthAI_Project/processed_data/images/xrays"
OUTPUT_DIR = "/content/drive/MyDrive/HealthAI_Project/processed_data"
os.makedirs(OUTPUT_DIR, exist_ok=True)

IMG_SIZE = 224
SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)

print("‚úÖ Using data dir:", DATA_DIR)
print("TensorFlow version:", tf.__version__)

# ==============================================================
# 1. Preprocessing
# ==============================================================
images, labels = [], []
for filename in os.listdir(DATA_DIR):
    if filename.endswith(".png"):
        img_path = os.path.join(DATA_DIR, filename)
        label = 1 if "Pneumonia" in filename else 0
        images.append(img_path)
        labels.append(label)

combined = list(zip(images, labels))
random.shuffle(combined)
images, labels = zip(*combined)

X_train_paths, X_val_paths, y_train, y_val = train_test_split(
    images, labels, test_size=0.2, random_state=SEED, stratify=labels
)

def load_and_preprocess_image(image_path):
    img = Image.open(image_path).convert("RGB")
    img = img.resize((IMG_SIZE, IMG_SIZE))
    img = np.array(img).astype("float32") / 255.0
    return img

X_train = np.array([load_and_preprocess_image(p) for p in X_train_paths])
X_val   = np.array([load_and_preprocess_image(p) for p in X_val_paths])
y_train, y_val = np.array(y_train), np.array(y_val)

# Save arrays & CSVs
np.save(os.path.join(OUTPUT_DIR, "X_train.npy"), X_train)
np.save(os.path.join(OUTPUT_DIR, "y_train.npy"), y_train)
np.save(os.path.join(OUTPUT_DIR, "X_val.npy"), X_val)
np.save(os.path.join(OUTPUT_DIR, "y_val.npy"), y_val)

pd.DataFrame({"image_path": X_train_paths, "label": y_train}).to_csv(os.path.join(OUTPUT_DIR, "train_labels.csv"), index=False)
pd.DataFrame({"image_path": X_val_paths, "label": y_val}).to_csv(os.path.join(OUTPUT_DIR, "val_labels.csv"), index=False)

print("‚úÖ Preprocessing complete. Saved .npy and CSV files to:", OUTPUT_DIR)
print("X_train:", X_train.shape, "y_train:", y_train.shape)
print("X_val:", X_val.shape, "y_val:", y_val.shape)

# Sample images
fig, axes = plt.subplots(2, 5, figsize=(15,6))
for i, ax in enumerate(axes.ravel()):
    idx = random.randint(0, len(X_train)-1)
    ax.imshow(X_train[idx])
    label = "Pneumonia" if y_train[idx]==1 else "Normal"
    ax.set_title(label)
    ax.axis("off")
plt.suptitle("Sample Training Images (Normalized)", fontsize=16)
plt.show()

# Class distribution
counts = collections.Counter(y_train)
plt.bar(["Normal (0)", "Pneumonia (1)"], [counts[0], counts[1]])
plt.title("Training Class Distribution")
plt.show()

# ==============================================================
# 2. CNN Model Training (EfficientNetB0)
# ==============================================================
IMG_SHAPE = (IMG_SIZE, IMG_SIZE, 3)
BATCH_SIZE = 16
AUTOTUNE = tf.data.AUTOTUNE

train_ds = tf.data.Dataset.from_tensor_slices((X_train, y_train)).shuffle(1024, seed=SEED).batch(BATCH_SIZE).prefetch(AUTOTUNE)
val_ds   = tf.data.Dataset.from_tensor_slices((X_val, y_val)).batch(BATCH_SIZE).prefetch(AUTOTUNE)

base_model = tf.keras.applications.EfficientNetB0(include_top=False, weights='imagenet', input_shape=IMG_SHAPE)
base_model.trainable = False

inputs = layers.Input(shape=IMG_SHAPE)
x = tf.keras.applications.efficientnet.preprocess_input(inputs)
x = base_model(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(128, activation='relu')(x)
outputs = layers.Dense(1, activation='sigmoid')(x)
model = models.Model(inputs, outputs)

metrics = [
    tf.keras.metrics.BinaryAccuracy(name='accuracy'),
    tf.keras.metrics.Precision(name='precision'),
    tf.keras.metrics.Recall(name='recall'),
    tf.keras.metrics.AUC(name='auc'),
]

model.compile(optimizer=keras.optimizers.Adam(1e-4), loss='binary_crossentropy', metrics=metrics)
model.summary()

callbacks = [
    keras.callbacks.ModelCheckpoint(os.path.join(OUTPUT_DIR, 'cnn_xray_best.h5'), save_best_only=True, monitor='val_auc', mode='max'),
    keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3),
    keras.callbacks.EarlyStopping(monitor='val_auc', patience=6, mode='max', restore_best_weights=True)
]

history = model.fit(train_ds, validation_data=val_ds, epochs=5, callbacks=callbacks)
model.save(os.path.join(OUTPUT_DIR, 'cnn_xray_final.h5'))
print("‚úÖ Model training complete and saved.")

# ==============================================================
# 3. Evaluation
# ==============================================================
y_true, y_pred_probs = [], []
for xb, yb in val_ds.unbatch().batch(1024):
    preds = model.predict(xb, verbose=0)
    y_true.extend(yb.numpy().tolist())
    y_pred_probs.extend(preds.ravel().tolist())

y_true = np.array(y_true)
y_pred_probs = np.array(y_pred_probs)
y_pred = (y_pred_probs >= 0.5).astype(int)

print("AUC:", roc_auc_score(y_true, y_pred_probs))
print("\nClassification report:\n", classification_report(y_true, y_pred, target_names=['NORMAL','PNEUMONIA']))
print("\nConfusion matrix:\n", confusion_matrix(y_true, y_pred))

# Training plots
plt.figure(); plt.plot(history.history['loss']); plt.plot(history.history['val_loss']); plt.legend(['train','val']); plt.title("Loss"); plt.savefig(os.path.join(OUTPUT_DIR, "cnn_loss_plot.png")); plt.show()
plt.figure(); plt.plot(history.history['accuracy']); plt.plot(history.history['val_accuracy']); plt.plot(history.history['auc']); plt.plot(history.history['val_auc']); plt.legend(['train_acc','val_acc','train_auc','val_auc']); plt.title("Accuracy & AUC"); plt.savefig(os.path.join(OUTPUT_DIR, "cnn_acc_auc_plot.png")); plt.show()

# Save metrics
report = classification_report(y_true, y_pred, output_dict=True)
with open(os.path.join(OUTPUT_DIR, 'cnn_classification_report.json'), 'w') as f: json.dump(report, f, indent=2)
cm = confusion_matrix(y_true, y_pred)
pd.DataFrame(cm, index=['NORMAL','PNEUMONIA'], columns=['Pred_NORMAL','Pred_PNEUMONIA']).to_csv(os.path.join(OUTPUT_DIR, 'cnn_confusion_matrix.csv'))

# ==============================================================
# 4. Occlusion Sensitivity (Interpretability)
# ==============================================================
def occlusion_sensitivity(img_tensor, model, patch=32, stride=32):
    img = img_tensor[0].numpy()
    H, W = img.shape[:2]
    base_pred = float(model.predict(img_tensor)[0,0])
    heatmap = np.zeros((H, W), dtype=np.float32)
    counts = np.zeros_like(heatmap)

    for y in range(0, H, stride):
        for x in range(0, W, stride):
            y2 = min(H, y+patch)
            x2 = min(W, x+patch)
            img_occ = img.copy()
            img_occ[y:y2, x:x2, :] = 0.0
            p = float(model.predict(img_occ[np.newaxis,...])[0,0])
            heatmap[y:y2, x:x2] += (base_pred - p)
            counts[y:y2, x:x2] += 1

    heatmap /= np.maximum(counts, 1e-8)
    heatmap = (heatmap - heatmap.min()) / (heatmap.max() - heatmap.min() + 1e-8)
    return heatmap

# Show occlusion maps for a few samples
samples = []
for xb, yb in val_ds.take(1):
    for i in range(min(3, xb.shape[0])):
        samples.append((xb[i:i+1], int(yb[i].numpy())))

fig = plt.figure(figsize=(12,6))
for i,(img_tensor,true_lbl) in enumerate(samples,1):
    heatmap = occlusion_sensitivity(img_tensor, model)
    img = (img_tensor[0].numpy()*255).astype(np.uint8)
    cmap = plt.get_cmap("jet")
    heatmap_color = (cmap(heatmap)[:,:,:3]*255).astype(np.uint8)
    overlay = (0.6*img + 0.4*heatmap_color).astype(np.uint8)
    ax = fig.add_subplot(1,3,i)
    ax.imshow(overlay)
    ax.axis("off")
    ax.set_title("True: " + ("PNEUMONIA" if true_lbl==1 else "NORMAL"))
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR,'cnn_occlusion_examples.png'))
plt.show()

print("‚úÖ All outputs (metrics, plots, occlusion maps) saved in:", OUTPUT_DIR)
```

```python
# ==============================================================
# 2. CNN Model Training (Fine-tuned EfficientNetB0 with Augmentation + Class Weights)
# ==============================================================

IMG_SHAPE = (IMG_SIZE, IMG_SIZE, 3)
BATCH_SIZE = 16
AUTOTUNE = tf.data.AUTOTUNE

# Data pipeline
train_ds = tf.data.Dataset.from_tensor_slices((X_train, y_train)) \
    .shuffle(1024, seed=SEED).batch(BATCH_SIZE).prefetch(AUTOTUNE)
val_ds   = tf.data.Dataset.from_tensor_slices((X_val, y_val)) \
    .batch(BATCH_SIZE).prefetch(AUTOTUNE)

# Data augmentation
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
])

# Base model
base_model = tf.keras.applications.EfficientNetB0(include_top=False, weights='imagenet', input_shape=IMG_SHAPE)
base_model.trainable = True
for layer in base_model.layers[:-20]:  # freeze most, unfreeze last 20
    layer.trainable = False

# Model architecture
inputs = layers.Input(shape=IMG_SHAPE)
x = data_augmentation(inputs)
x = tf.keras.applications.efficientnet.preprocess_input(x)
x = base_model(x, training=True)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.3)(x)
x = layers.Dense(128, activation='relu')(x)
outputs = layers.Dense(1, activation='sigmoid')(x)
model = models.Model(inputs, outputs)

# Compile with smaller LR
metrics = [
    tf.keras.metrics.BinaryAccuracy(name='accuracy'),
    tf.keras.metrics.Precision(name='precision'),
    tf.keras.metrics.Recall(name='recall'),
    tf.keras.metrics.AUC(name='auc'),
]
model.compile(optimizer=keras.optimizers.Adam(1e-5), loss='binary_crossentropy', metrics=metrics)

# Compute class weights
from sklearn.utils.class_weight import compute_class_weight
class_weights = compute_class_weight("balanced", classes=np.unique(y_train), y=y_train)
class_weights = dict(enumerate(class_weights))
print("Class weights:", class_weights)

# Callbacks
callbacks = [
    keras.callbacks.ModelCheckpoint(os.path.join(OUTPUT_DIR, 'cnn_xray_best.h5'), save_best_only=True, monitor='val_auc', mode='max'),
    keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3),
    keras.callbacks.EarlyStopping(monitor='val_auc', patience=6, mode='max', restore_best_weights=True)
]

# Train longer
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=7,
    callbacks=callbacks,
    class_weight=class_weights
)

model.save(os.path.join(OUTPUT_DIR, 'cnn_xray_final.h5'))
print("‚úÖ Fine-tuned model training complete and saved.")
```

**8_lstm_timeseries.ipynb ‚Üí Sequence modeling on vitals**

```python
# --- Imports ---
import os, random, json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (classification_report, confusion_matrix,
                             roc_auc_score, recall_score, mean_squared_error)

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models

# --- Reproducibility ---
SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)

# --- Synthetic vitals dataset ---
# Features: [HR, BP, Resp, Temp]
N, T, F = 1200, 30, 4   # more samples for stability
X = np.zeros((N, T, F))
y_class = np.zeros(N)

for i in range(N):
    hr = 60 + np.random.randn(T).cumsum()/10
    bp = 120 + np.random.randn(T).cumsum()/10
    resp = 16 + np.random.randn(T).cumsum()/50
    temp = 37 + np.random.randn(T).cumsum()/100
    seq = np.vstack([hr, bp, resp, temp]).T
    X[i] = seq
    # deterioration if HR > 100 or BP > 140 at last timestep
    y_class[i] = 1 if (hr[-1] > 100 or bp[-1] > 140) else 0

print("Class distribution:", np.bincount(y_class.astype(int)))

# --- Scale features ---
scaler = StandardScaler()
X_reshaped = X.reshape(-1, F)
X_scaled = scaler.fit_transform(X_reshaped).reshape(N, T, F)

# --- Split train/val (stratified ensures both classes appear) ---
X_train, X_val, y_train, y_val = train_test_split(
    X_scaled, y_class, test_size=0.2, random_state=SEED, stratify=y_class
)

# ==============================================================
# Track A: Classification ‚Üí Early Warning System
# ==============================================================

clf_model = models.Sequential([
    layers.Input(shape=(T, F)),
    layers.LSTM(64, return_sequences=True),
    layers.Dropout(0.3),
    layers.LSTM(32),
    layers.Dense(1, activation="sigmoid")
])

clf_model.compile(optimizer="adam", loss="binary_crossentropy",
                  metrics=[tf.keras.metrics.BinaryAccuracy(name="accuracy"),
                           tf.keras.metrics.AUC(name="auc")])

history_clf = clf_model.fit(X_train, y_train, validation_data=(X_val, y_val),
                            epochs=10, batch_size=32, verbose=1)

# --- Evaluation ---
y_pred_probs = clf_model.predict(X_val).ravel()
y_pred = (y_pred_probs >= 0.5).astype(int)

print("\nClassification Report:\n",
      classification_report(y_val, y_pred, labels=[0,1], target_names=["Normal","Deterioration"], zero_division=0))
print("AUC:", roc_auc_score(y_val, y_pred_probs))
print("Confusion Matrix:\n", confusion_matrix(y_val, y_pred))

# Early Warning Detection Rate = Recall for positive class
ewdr = recall_score(y_val, y_pred, pos_label=1)
print("Early Warning Detection Rate:", ewdr)

# ==============================================================
# Track B: Forecasting ‚Üí Next-step vitals prediction
# ==============================================================

# Prepare regression data
X_reg = X_scaled[:, :-1, :]   # first 29 timesteps
y_reg = X_scaled[:, -1, :]    # last timestep vitals (all 4)

Xr_train, Xr_val, yr_train, yr_val = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=SEED
)

reg_model = models.Sequential([
    layers.Input(shape=(T-1, F)),
    layers.LSTM(64, return_sequences=True),
    layers.Dropout(0.3),
    layers.LSTM(32),
    layers.Dense(F)   # predict 4 vitals
])

reg_model.compile(optimizer="adam", loss="mse")
history_reg = reg_model.fit(Xr_train, yr_train, validation_data=(Xr_val, yr_val),
                            epochs=15, batch_size=32, verbose=1)

# --- Evaluation ---
y_pred_reg = reg_model.predict(Xr_val)

# RMSE per feature
rmse_per_feature = np.sqrt(((yr_val - y_pred_reg) ** 2).mean(axis=0))
for i, name in enumerate(["HR","BP","Resp","Temp"]):
    print(f"RMSE {name}: {rmse_per_feature[i]:.4f}")

rmse_all = np.sqrt(mean_squared_error(yr_val, y_pred_reg))
print("Overall Forecast RMSE:", rmse_all)

# --- Forecast plots for all vitals ---
vital_names = ["Heart Rate (HR)", "Blood Pressure (BP)", "Respiration (Resp)", "Temperature (Temp)"]
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()

for i in range(4):
    axes[i].plot(yr_val[:50, i], label="True", marker="o")
    axes[i].plot(y_pred_reg[:50, i], label="Pred", marker="x")
    axes[i].set_title(vital_names[i])
    axes[i].legend()

plt.suptitle("LSTM Forecasts for Vitals (First 50 Samples)", fontsize=14)
plt.tight_layout()
plt.show()

print("‚úÖ Classification (EWDR) + Regression (RMSE per vital + plots) complete")
```

| Aspect                | Track A+B (Separate)           | Multi-Output LSTM                           |
| --------------------- | ------------------------------ | ------------------------------------------- |
| Number of models      | 2 (classification, regression) | 1 (two heads)                               |
| Training              | Separate training runs         | Joint training                              |
| Loss handling         | Independent losses             | Multi-task loss balancing (can add weights) |
| Efficiency            | Slower (twice training)        | Faster (single model)                       |
| Interpretability      | Easier to debug per task       | More entangled                              |
| Project guideline fit | ‚úÖ Still valid                  | ‚úÖ More elegant                              |

```python
# ==============================================================
# Multi-Output LSTM ‚Üí Forecast vitals + Early warning detection
# ==============================================================

import os, random, json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, classification_report, confusion_matrix, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras import backend as K

# --- Paths ---
OUTPUT_DIR = "/content/drive/MyDrive/HealthAI_Project/processed_data"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# --- Seed ---
SEED = 42
np.random.seed(SEED); tf.random.set_seed(SEED)

# ==============================================================
# 1. Synthetic Example Dataset (Replace with your vitals data)
# ==============================================================
# Suppose we have HR (heart rate) as time series & deterioration labels
time_steps = 50   # sequence length
features = 1      # HR only (can add BP, SpO2, etc.)
samples = 1500    # number of patients

# Generate synthetic vitals
X = np.sin(np.linspace(0,100,samples*time_steps)).reshape(samples, time_steps, features)
X += np.random.normal(0,0.1,X.shape)

# Forecast target = next HR value
y_forecast = np.roll(X[:,:,0], -1, axis=1)[:, -1]  # next step HR

# Classification target = deterioration risk (1 if HR > 0.5 in last step)
y_class = (X[:,-1,0] > 0.5).astype(int)

# Train/val split
X_train, X_val, y_fore_train, y_fore_val, y_cls_train, y_cls_val = train_test_split(
    X, y_forecast, y_class, test_size=0.2, random_state=SEED, stratify=y_class
)

print("‚úÖ Data prepared:", X_train.shape, y_fore_train.shape, y_cls_train.shape)

# ==============================================================
# 2. Multi-output LSTM Model
# ==============================================================

inputs = layers.Input(shape=(time_steps, features))
x = layers.LSTM(64, return_sequences=True)(inputs)
x = layers.LSTM(32)(x)

# Regression head ‚Üí forecast vital
out_forecast = layers.Dense(1, name="forecast")(x)

# Classification head ‚Üí early warning
out_class = layers.Dense(1, activation="sigmoid", name="classify")(x)

model = models.Model(inputs, [out_forecast, out_class])

# ==============================================================
# Custom Weighted Loss Function for Classification
# ==============================================================

# Compute class weights for the classification task
class_weights = compute_class_weight('balanced', classes=np.unique(y_cls_train), y=y_cls_train)
class_weights = dict(enumerate(class_weights))
print("Class weights for classification:", class_weights)

def weighted_binary_crossentropy(y_true, y_pred):
    # Cast to float to avoid potential type issues
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)

    # Calculate weights based on true labels
    weights = tf.gather(tf.constant([class_weights[0], class_weights[1]]), tf.cast(y_true, tf.int32))

    # Calculate binary crossentropy
    bce = K.binary_crossentropy(y_true, y_pred)

    # Apply weights
    weighted_bce = bce * weights

    return K.mean(weighted_bce) # Return mean loss

# ==============================================================
# Compile with custom loss for classification
# ==============================================================

model.compile(
    optimizer=tf.keras.optimizers.Adam(1e-3),
    loss={"forecast":"mse", "classify": weighted_binary_crossentropy},
    metrics={"forecast":["mse"], "classify":["accuracy","AUC","Precision","Recall"]}
)

model.summary()

# ==============================================================
# 3. Train
# ==============================================================

history = model.fit(
    X_train, {"forecast": y_fore_train, "classify": y_cls_train},
    validation_data=(X_val, {"forecast": y_fore_val, "classify": y_cls_val}),
    epochs=10, batch_size=32,
    # Removed class_weight from fit as it's now handled in the custom loss
)

# ==============================================================
# 4. Evaluation
# ==============================================================

# Forecast evaluation
y_pred_fore, y_pred_cls = model.predict(X_val)
rmse = np.sqrt(mean_squared_error(y_fore_val, y_pred_fore))
print("Forecast RMSE:", rmse)

# Classification evaluation
y_pred_cls_bin = (y_pred_cls >= 0.5).astype(int)
report = classification_report(y_cls_val, y_pred_cls_bin, target_names=["Normal","Deterioration"], output_dict=True)
print("\nClassification Report:\n", classification_report(y_cls_val, y_pred_cls_bin, target_names=["Normal","Deterioration"]))
print("Confusion Matrix:\n", confusion_matrix(y_cls_val, y_pred_cls_bin))
print("AUC:", roc_auc_score(y_cls_val, y_pred_cls))

# Early warning detection rate = Recall for Deterioration
ewdr = report["Deterioration"]["recall"] # Corrected key
print("Early Warning Detection Rate (EWDR):", ewdr)

# ==============================================================
# 5. Visualization
# ==============================================================

# Plot forecast for some patients
plt.figure(figsize=(10,5))
plt.plot(y_fore_val[:100], label="True HR")
plt.plot(y_pred_fore[:100], label="Pred HR")
plt.legend(); plt.title("LSTM HR Forecast")
plt.savefig(os.path.join(OUTPUT_DIR, "lstm_forecast.png"))
plt.show()

# Save metrics
results = {
    "Forecast_RMSE": float(rmse),
    "AUC": float(roc_auc_score(y_cls_val, y_pred_cls)),
    "EWDR": float(ewdr)
}
with open(os.path.join(OUTPUT_DIR,"lstm_results.json"),"w") as f: json.dump(results,f,indent=2)

print("‚úÖ All outputs (forecast plot, metrics) saved to:", OUTPUT_DIR)
```

**9_biobert_nlp.ipynb ‚Üí BioBERT/ClinicalBERT for clinical text**

```python
# ==============================================================
# Single-cell: Bio/ClinicalBERT NER (bc5cdr) + English<->Tamil translation (M2M100)
# Produces: NER metrics (seqeval), translation metrics (BLEU, COMET if available)
# Saves: fine-tuned NER model + tokenizer + metrics JSON to Drive
# Notes: Disables W&B auto-logging. COMET is optional (may be slow).
# ==============================================================
# Install dependencies (first run in Colab)
!pip install -q transformers datasets evaluate sacrebleu sentencepiece sacremoses

import os, json, random, time, sys, traceback
from pathlib import Path
import numpy as np
import torch
import evaluate
from datasets import load_dataset, DatasetDict
from transformers import (
    AutoTokenizer, AutoModelForTokenClassification,
    DataCollatorForTokenClassification, Trainer, TrainingArguments,
    M2M100ForConditionalGeneration, M2M100Tokenizer
)

# -----------------------------
# Configuration
# -----------------------------
OUTPUT_DIR = Path("/content/drive/MyDrive/HealthAI_Project/processed_data/biobert_ner_en_ta")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# Choose model for token-classification (Bio/ClinicalBERT). Swap if you prefer another checkpoint.
NER_MODEL = "emilyalsentzer/Bio_ClinicalBERT"   # ClinicalBERT (good for clinical notes)
NER_MAX_LEN = 128
NER_BATCH = 8
NER_EPOCHS = 3
SEED = 42

# Translation model (M2M100 supports Tamil reliably). This is large (~1.6GB); Colab GPU recommended.
MT_MODEL = "facebook/m2m100_418M"
MT_BATCH = 32
MT_MAX_LEN = 128

# Disable WandB to avoid interactive prompt
os.environ["WANDB_DISABLED"] = "true"

# Reproducibility
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
if torch.cuda.is_available(): torch.cuda.manual_seed_all(SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# -----------------------------
# 1) LOAD & PREPARE NER DATA (bc5cdr)
# -----------------------------
print("\n--> Loading ade_corpus_v2 dataset for biomedical NER (disease/chemical)...")
# Assuming "ade_corpus_v2" "Ade_corpus_v2_classification" has 'text' and 'label'
ds = load_dataset("ade_corpus_v2", "Ade_corpus_v2_classification")

# Ensure we have train/validation/test splits
if "test" not in ds:
    # Split train into train and temp (which will be split into validation and test)
    train_temp_split = ds["train"].train_test_split(test_size=0.2, seed=SEED)
    train_ds = train_temp_split["train"]
    temp_ds = train_temp_split["test"]

    # Split temp into validation and test
    val_test_split = temp_ds.train_test_split(test_size=0.5, seed=SEED)
    val_ds = val_test_split["train"]
    test_ds = val_test_split["test"]

    ds = DatasetDict({"train": train_ds, "validation": val_ds, "test": test_ds})

elif "validation" not in ds:
    # make small validation split from train, use existing test
    ds_train_val = ds["train"].train_test_split(test_size=0.1, seed=SEED)
    ds = DatasetDict({"train": ds_train_val["train"], "validation": ds_train_val["test"], "test": ds["test"]})
else:
    # Dataset already has train, validation, and test splits
    pass


print("Splits sizes:", {k: len(ds[k]) for k in ds.keys()})
features = ds["train"].features
# Check if the dataset has 'text' and 'label' for classification
if "text" not in features or "label" not in features:
     raise RuntimeError("Dataset missing 'text' or 'label' features. Choose a proper classification dataset.")

# This code was originally for NER, but the dataset loaded is for classification.
# We will adapt the translation part to use the 'text' column.
# The NER part will not run correctly with this dataset without significant changes.
print("Dataset has 'text' and 'label' features suitable for classification, not NER.")

# -----------------------------
# 2) Tokenizer + Prepare for Translation
# -----------------------------
print("\n--> Preparing data for Translation...")

# Use the 'text' column for translation samples
def extract_sentences_from_split(split_ds, n=200):
    out = []
    # Assuming split_ds contains examples with a 'text' key
    for ex in split_ds:
        s = ""
        if "text" in ex and isinstance(ex["text"], str):
            s = ex["text"]
        elif isinstance(ex, str): # Handle case where example itself is a string
             s = ex

        if len(s.strip()) > 5:
            out.append(s)
        if len(out) >= n:
            break
    return out


test_src = extract_sentences_from_split(ds["test"], n=120)
if len(test_src)==0:
    # fallback synthetic
    test_src = ["Patient has fever and cough.", "Discharged stable."]


# -----------------------------
# 3) Translation: English <-> Tamil (M2M100)
# -----------------------------
print("\n--> Translation: EN <-> TA using facebook/m2m100_418M")

mt_ok = True
try:
    mt_tokenizer = M2M100Tokenizer.from_pretrained(MT_MODEL)
    mt_model = M2M100ForConditionalGeneration.from_pretrained(MT_MODEL).to(device)
except Exception as e:
    print("Failed to load M2M100 model (may be large). Exception:\n", e)
    mt_ok = False

# We'll use a smaller subset for speed
test_src = test_src[:60]

translation_results = {"en_ta": None, "ta_en": None, "bleu_en_ta_back": None, "comet": None}
sacrebleu = evaluate.load("sacrebleu")

if mt_ok:
    # EN -> TA
    print(f"\nTranslating {len(test_src)} sentences EN -> TA ...")
    en_sentences = test_src
    # tokenize & generate in batches
    en_to_ta = []
    batch = []
    for i, s in enumerate(en_sentences):
        batch.append(s)
        if len(batch)==MT_BATCH or i==len(en_sentences)-1:
            encoded = mt_tokenizer(batch, return_tensors="pt", padding=True, truncation=True, max_length=MT_MAX_LEN).to(device)
            # force target language to Tamil
            ta_id = mt_tokenizer.get_lang_id("ta") if "ta" in mt_tokenizer.lang_code_to_id else None
            if ta_id is not None:
                outputs = mt_model.generate(**encoded, forced_bos_token_id=ta_id, max_length=MT_MAX_LEN)
            else:
                outputs = mt_model.generate(**encoded, max_length=MT_MAX_LEN)
            decoded = mt_tokenizer.batch_decode(outputs, skip_special_tokens=True)
            en_to_ta.extend(decoded)
            batch = []
    translation_results["en_ta"] = en_to_ta

    # TA -> EN (back-translate)
    print(f"\nBack-translating {len(en_to_ta)} sentences TA -> EN ...")
    ta_to_en = []
    batch = []
    for i, s in enumerate(en_to_ta):
        batch.append(s)
        if len(batch)==MT_BATCH or i==len(en_to_ta)-1: # Corrected variable name
            encoded = mt_tokenizer(batch, return_tensors="pt", padding=True, truncation=True, max_length=MT_MAX_LEN).to(device)
            en_id = mt_tokenizer.get_lang_id("en") if "en" in mt_tokenizer.lang_code_to_id else None
            if en_id is not None:
                outputs = mt_model.generate(**encoded, forced_bos_token_id=en_id, max_length=MT_MAX_LEN)
            else:
                outputs = mt_model.generate(**encoded, max_length=MT_MAX_LEN)
            decoded = mt_tokenizer.batch_decode(outputs, skip_special_tokens=True)
            ta_to_en.extend(decoded)
            batch = []
    translation_results["ta_en_back"] = ta_to_en

    # Evaluate BLEU of back-translation (how close back-translated -> original)
    print("\nComputing sacreBLEU for back-translated text (ta->en back) ...")
    bleu = sacrebleu.compute(predictions=ta_to_en, references=[[s] for s in en_sentences])
    translation_results["bleu_en_ta_back"] = bleu
    print("sacreBLEU (EN <- TA back):", bleu)

    # Try COMET (best-effort ‚Äî may be slow)
    try:
        print("\nAttempting COMET scoring (may download models; could take time)...")
        comet = evaluate.load("comet")
        # COMET wants lists of predictions/references and typically sources; we'll provide:
        comet_scores = comet.compute(predictions=ta_to_en, references=en_sentences, sources=en_sentences) # Added sources
        translation_results["comet"] = comet_scores
        print("COMET result (sample):", {k: translation_results["comet"].get(k) for k in translation_results["comet"] if k in ["score"]})
    except Exception as e:
        print("COMET unavailable or failed (falling back). Exception:", e)
        translation_results["comet_error"] = str(e)
else:
    print("\nSkipping translation (M2M100 unavailable).")

# Save translation outputs
with open(OUTPUT_DIR / "translation_results.json", "w") as f:
    json.dump(translation_results, f, indent=2)

# -----------------------------
# 5) Final save & summary
# -----------------------------
# NER metrics will not be available with this dataset
ner_metrics = {"error": "NER part skipped due to incompatible dataset"}

summary = {
    "ner_metrics": ner_metrics,
    "translation_summary": {
        "mt_loaded": mt_ok,
        "n_test_sentences": len(test_src),
        "bleu_en_ta_back": translation_results.get("bleu_en_ta_back"),
        "comet": translation_results.get("comet") or translation_results.get("comet_error")
    }
}
with open(OUTPUT_DIR / "pipeline_summary.json", "w") as f:
    json.dump(summary, f, indent=2)

print("\nAll done. Outputs saved to:", OUTPUT_DIR)
# Adjusted summary printout
print("Summary (high level):")
print(" - NER (seqeval):", ner_metrics)
print(" - MT loaded:", mt_ok)
print(" - BLEU (EN <- TA back):", translation_results.get("bleu_en_ta_back"))
print(" - COMET:", "available" if translation_results.get("comet") else translation_results.get("comet_error", "not computed"))
```

```python
# ==============================================================
# Bio/ClinicalBERT NER (synthetic) + English<->Tamil Translation (M2M100)
# Produces: NER metrics (F1 + confusion matrix), BLEU, COMET
# Saves: fine-tuned model, metrics, plots, sample translations to Drive
# ==============================================================

!pip install -q transformers datasets evaluate sacrebleu sentencepiece sacremoses seaborn seqeval

import os, json, random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

import torch
from datasets import Dataset, DatasetDict
import evaluate
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

from transformers import (
    AutoTokenizer, AutoModelForTokenClassification,
    DataCollatorForTokenClassification, Trainer, TrainingArguments,
    M2M100ForConditionalGeneration, M2M100Tokenizer
)

# -----------------------------
# Config & Paths
# -----------------------------
OUTPUT_DIR = Path("/content/drive/MyDrive/HealthAI_Project/processed_data/biobert_ner_en_ta")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

NER_MODEL = "emilyalsentzer/Bio_ClinicalBERT"
MT_MODEL = "facebook/m2m100_418M"

# -----------------------------
# 1) Synthetic NER Dataset (100 sentences, 3 diseases) with IOB tagging
# -----------------------------
entities = ["diabetes", "pneumonia", "cancer"]
sentences, tags = [], []

for i in range(100):
    disease = random.choice(entities)
    template = random.choice([
        f"Patient has {disease} and requires treatment.",
        f"Symptoms indicate possible {disease} diagnosis.",
        f"No signs of {disease} found in patient record.",
        f"Family history shows risk of {disease}.",
        f"{disease.capitalize()} management plan was initiated."
    ])
    words = template.split()
    ner_tags = []
    for w in words:
        if disease in w.lower():
            # Simplified IOB: assuming single-word disease mentions for synthetic data
            ner_tags.append("B-DISEASE")
        else:
            ner_tags.append("O")
    sentences.append(words)
    tags.append(ner_tags)

# Updated labels for IOB tagging
labels = ["O", "B-DISEASE"]
# Correctly create label2id and id2label
label2id = {label: i for i, label in enumerate(labels)}
id2label = {i: label for i, label in enumerate(labels)}

tag_ids = [[label2id[t] for t in seq] for seq in tags]

# Split train/val/test
train_size, val_size = 70, 15
train_ds = Dataset.from_dict({"tokens": sentences[:train_size], "ner_tags": tag_ids[:train_size]})
val_ds   = Dataset.from_dict({"tokens": sentences[train_size:train_size+val_size], "ner_tags": tag_ids[train_size:train_size+val_size]})
test_ds  = Dataset.from_dict({"tokens": sentences[train_size+val_size:], "ner_tags": tag_ids[train_size+val_size:]})
ds = DatasetDict({"train": train_ds, "validation": val_ds, "test": test_ds})

print("Synthetic dataset:", ds)

# -----------------------------
# 2) Tokenize + Align Labels
# -----------------------------
tokenizer = AutoTokenizer.from_pretrained(NER_MODEL, use_fast=True)
def tokenize_and_align_labels(examples):
    tokenized = tokenizer(examples["tokens"], truncation=True, is_split_into_words=True,
                          padding="max_length", max_length=64)
    all_labels = []
    for i, labels in enumerate(examples["ner_tags"]):
        word_ids = tokenized.word_ids(batch_index=i)
        label_ids, prev = [], None
        for word_idx in word_ids:
            if word_idx is None: label_ids.append(-100)
            elif word_idx != prev: label_ids.append(labels[word_idx])
            else: label_ids.append(-100)
            prev = word_idx
        all_labels.append(label_ids)
    tokenized["labels"] = all_labels
    return tokenized

tokenized = ds.map(tokenize_and_align_labels, batched=True, remove_columns=ds["train"].column_names)

# -----------------------------
# 3) Train BioBERT for NER
# -----------------------------
# Load the pre-trained model without specifying num_labels, id2label, or label2id initially
model = AutoModelForTokenClassification.from_pretrained(NER_MODEL).to(device)

# Explicitly re-initialize the classifier head for the correct number of labels
# The classifier is typically a linear layer named 'classifier' or 'cls' depending on the model architecture
# Check the model summary or documentation to confirm the layer name
if hasattr(model, 'classifier'):
    classifier_layer_name = 'classifier'
elif hasattr(model, 'cls'):
    classifier_layer_name = 'cls'
else:
    raise AttributeError("Could not find classifier layer in the model. Check model architecture.")

# Get the number of input features to the classifier layer
classifier_input_features = getattr(model, classifier_layer_name).in_features

# Replace the classifier layer with a new one for the correct number of labels
setattr(model, classifier_layer_name, torch.nn.Linear(classifier_input_features, len(labels)).to(device))

# Set the id2label and label2id attributes on the model for proper saving and loading later
model.config.id2label = id2label
model.config.label2id = label2id

print(f"\n--> Loaded model for token classification: {NER_MODEL}")
print(f"Re-initialized classifier head for {len(labels)} labels.")


data_collator = DataCollatorForTokenClassification(tokenizer)
seqeval = evaluate.load("seqeval")

def compute_metrics(p):
    logits, labels_arr = p
    preds = np.argmax(logits, axis=2)
    true_labels, true_preds = [], []
    for i in range(labels_arr.shape[0]):
        lab_seq, pred_seq = [], []
        for j in range(labels_arr.shape[1]):
            # Only consider positions that are not padding/special tokens (-100)
            if labels_arr[i, j] != -100:
                # Append true label
                if labels_arr[i, j] in id2label:
                    lab_seq.append(id2label[labels_arr[i, j]])
                else:
                    # Handle unexpected true label IDs by treating them as O
                    lab_seq.append("O")

                # Append predicted label
                if preds[i, j] in id2label:
                    pred_seq.append(id2label[preds[i, j]])
                else:
                    # Handle unexpected predicted label IDs by treating them as O
                    pred_seq.append("O")

        true_labels.append(lab_seq)
        true_preds.append(pred_seq)

    # seqeval correctly handles IOB tags
    return seqeval.compute(predictions=true_preds, references=true_labels)

args = TrainingArguments(
    output_dir=str(OUTPUT_DIR/"ner_ckpt"),
    eval_strategy="epoch", # Corrected argument name
    save_strategy="epoch",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    learning_rate=3e-5,
    weight_decay=0.01,
    logging_steps=10,
    report_to="none",
    seed=SEED
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=tokenized["train"],
    eval_dataset=tokenized["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

print("\nStarting NER fine-tuning... (this will take some minutes depending on dataset size & GPU)")
trainer.train()
ner_metrics = trainer.evaluate(tokenized["test"])
print("\nNER metrics:", ner_metrics)

# Confusion matrix (test set, flattened labels)
preds = trainer.predict(tokenized["test"]).predictions
preds = np.argmax(preds, axis=2).flatten()
labels_flat = np.array(tokenized["test"]["labels"]).flatten()
mask = labels_flat != -100
# Filter predictions to match the masked labels
preds_masked = preds[np.where(labels_flat != -100)[0]] # Filter preds where true label is not -100
cm = confusion_matrix(labels_flat[mask], preds_masked) # Use filtered preds
# Use the updated labels for display
disp = ConfusionMatrixDisplay(cm, display_labels=labels)
disp.plot(cmap="Blues")
plt.title("NER Confusion Matrix (IOB)")
plt.savefig(OUTPUT_DIR/"ner_confusion_matrix_iob.png") # Save with new name
plt.show()

with open(OUTPUT_DIR/"ner_metrics_iob.json","w") as f: json.dump(ner_metrics,f,indent=2) # Save metrics with new name

# -----------------------------
# 4) Translation EN <-> TA
# -----------------------------
mt_tokenizer = M2M100Tokenizer.from_pretrained(MT_MODEL)
mt_model = M2M100ForConditionalGeneration.from_pretrained(MT_MODEL).to(device)

sample_texts = [
    "Patient has diabetes and fever.",
    "No signs of pneumonia.",
    "Cancer treatment plan initiated."
]

def translate(texts, src_lang, tgt_lang):
    mt_tokenizer.src_lang = src_lang
    encoded = mt_tokenizer(texts, return_tensors="pt", padding=True, truncation=True).to(device)
    forced_id = mt_tokenizer.get_lang_id(tgt_lang)
    outputs = mt_model.generate(**encoded, forced_bos_token_id=forced_id, max_length=64)
    return mt_tokenizer.batch_decode(outputs, skip_special_tokens=True)

en_to_ta = translate(sample_texts, "en", "ta")
ta_to_en = translate(en_to_ta, "ta", "en")

# BLEU on back-translation
sacrebleu = evaluate.load("sacrebleu")
bleu = sacrebleu.compute(predictions=ta_to_en, references=[[s] for s in sample_texts])

# COMET (optional, may be slow)
try:
    comet = evaluate.load("comet")
    comet_scores = comet.compute(predictions=ta_to_en, references=sample_texts, sources=sample_texts)
except:
    comet_scores = {"error": "COMET not available"}

print("\nTranslation Example:")
print("EN:", sample_texts[0])
print("TA:", en_to_ta[0])
print("Back EN:", ta_to_en[0])

translation_results = {"BLEU": bleu, "COMET": comet_scores}
with open(OUTPUT_DIR/"translation_results.json","w") as f: json.dump(translation_results,f,indent=2)

# -----------------------------
# 5) Final Summary
# -----------------------------
summary = {
    "NER_F1": ner_metrics["eval_overall_f1"],
    "NER_precision": ner_metrics["eval_overall_precision"],
    "NER_recall": ner_metrics["eval_overall_recall"],
    "BLEU": bleu,
    "COMET": comet_scores
}
with open(OUTPUT_DIR/"pipeline_summary_iob.json","w") as f: json.dump(summary,f,indent=2) # Save summary with new name

print("\n‚úÖ All done. Outputs saved to:", OUTPUT_DIR)
```

```python
# ==============================================================
# Sentiment Analysis ‚Üí Fine-tuned BERT on patient feedback
# Produces: Precision, Recall, F1, MCC, Confusion Matrix
# Saves: trained model + metrics to Drive
# ==============================================================

!pip install -q transformers datasets evaluate seaborn scikit-learn

import os, json, random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

import torch
from datasets import Dataset, DatasetDict
from sklearn.metrics import classification_report, confusion_matrix, matthews_corrcoef
from transformers import (
    AutoTokenizer, AutoModelForSequenceClassification,
    TrainingArguments, Trainer, DataCollatorWithPadding
)

# -----------------------------
# Config & Paths
# -----------------------------
OUTPUT_DIR = Path("/content/drive/MyDrive/HealthAI_Project/processed_data/sentiment_analysis")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

MODEL_NAME = "bert-base-uncased"

# -----------------------------
# 1) Synthetic Sentiment Dataset
# -----------------------------
pos_feedback = [
    "The doctor was very kind and explained everything clearly.",
    "Great hospital experience, staff were caring and professional.",
    "I am satisfied with the treatment and the nurses were amazing."
]
neg_feedback = [
    "The waiting time was terrible and staff were rude.",
    "Very poor service, I felt ignored by the doctor.",
    "I am not satisfied, the treatment was ineffective."
]
neu_feedback = [
    "The hospital is okay, nothing special.",
    "Average experience, neither good nor bad.",
    "The facilities are fine but could be improved."
]

texts, labels = [], []
for _ in range(150):  # 150 synthetic samples * 3 = 450
    texts.append(random.choice(pos_feedback)); labels.append(2)  # positive
    texts.append(random.choice(neg_feedback)); labels.append(0)  # negative
    texts.append(random.choice(neu_feedback)); labels.append(1)  # neutral

# Split train/val
split_idx = int(0.8 * len(texts))
train_texts, val_texts = texts[:split_idx], texts[split_idx:]
train_labels, val_labels = labels[:split_idx], labels[split_idx:]

train_ds = Dataset.from_dict({"text": train_texts, "label": train_labels})
val_ds   = Dataset.from_dict({"text": val_texts, "label": val_labels})
ds = DatasetDict({"train": train_ds, "validation": val_ds})
print("Synthetic sentiment dataset:", ds)

# -----------------------------
# 2) Tokenize
# -----------------------------
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
def tokenize(batch): return tokenizer(batch["text"], truncation=True, padding="max_length", max_length=64)
ds_enc = ds.map(tokenize, batched=True)

# -----------------------------
# 3) Train BERT Classifier
# -----------------------------
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=3).to(device)
collator = DataCollatorWithPadding(tokenizer=tokenizer)

args = TrainingArguments(
    output_dir=str(OUTPUT_DIR/"bert_sentiment"),
    eval_strategy="epoch", # Corrected argument name
    save_strategy="epoch",
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    learning_rate=3e-5,
    weight_decay=0.01,
    logging_steps=10,
    report_to="none",
    seed=SEED,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    greater_is_better=True
)

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)
    report = classification_report(labels, preds, output_dict=True, zero_division=0)
    mcc = matthews_corrcoef(labels, preds)
    return {
        "precision": report["weighted avg"]["precision"],
        "recall": report["weighted avg"]["recall"],
        "f1": report["weighted avg"]["f1-score"],
        "mcc": mcc
    }

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=ds_enc["train"],
    eval_dataset=ds_enc["validation"],
    tokenizer=tokenizer,
    data_collator=collator,
    compute_metrics=compute_metrics
)

print("\nStarting sentiment fine-tuning...")
trainer.train()
results = trainer.evaluate()
print("\nFinal metrics:", results)

# -----------------------------
# 4) Confusion Matrix
# -----------------------------
preds = np.argmax(trainer.predict(ds_enc["validation"]).predictions, axis=1)
true = np.array(val_labels)
cm = confusion_matrix(true, preds)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Negative","Neutral","Positive"],
            yticklabels=["Negative","Neutral","Positive"])
plt.title("Sentiment Confusion Matrix")
plt.savefig(OUTPUT_DIR/"confusion_matrix.png")
plt.show()

# Save metrics
with open(OUTPUT_DIR/"sentiment_metrics.json","w") as f: json.dump(results,f,indent=2)

print("\n‚úÖ Sentiment analysis complete. Outputs saved to:", OUTPUT_DIR)
```
